```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
data = pd.read_csv('D:/study docs/Excel R/Data science/assignment/Decision Tree/Datasets/Fraud_check.csv')
```

```python
data.head()
```

```python
data=pd.get_dummies(data,columns=['Undergrad','Marital.Status','Urban'])
```

```python
data
```

```python
data["TaxInc"] = pd.cut(data["Taxable.Income"], bins = [10002,30000,99620], labels = ["Risky", "Good"])
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data.iloc[:,-1] = labelencoder.fit_transform(data.iloc[:,-1])
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data.iloc[:,0:3] = scaler.fit_transform(data.iloc[:,0:3])
```

```python
data
```

```python
data.dtypes
```

```python
data['TaxInc']=data['TaxInc'].astype('category') 
```

```python
data.dtypes
```

```python
x = data.iloc[:,:-1]
y = data['TaxInc']
```

```python
y.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y ,test_size = 0.2)
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
preds = model.predict(x_test)  
pd.Series(preds).value_counts()
```

```python
pd.crosstab(y_test,preds)
```
